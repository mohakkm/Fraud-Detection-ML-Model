```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv('/content/creditcard.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates(inplace=True)
```

```python
scaler = StandardScaler()
df["Amount"] = scaler.fit_transform(df[["Amount"]])
```

```python
df.fillna(df.median(), inplace=True)
df.fillna("Unknown", inplace=True)
```

```python
X = df.drop(["Class", "Time"], axis=1)
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
smote = SMOTE(sampling_strategy=1.0, random_state=42)  # Fully balance the dataset
X_train, y_train = smote.fit_resample(X_train, y_train)
```

```python
rf = RandomForestClassifier(class_weight="balanced", random_state=42)
```

```python
param_grid = {"n_estimators": [100, 200],"max_depth": [10, 20],"min_samples_split": [2, 5],"min_samples_leaf": [1, 2]}
```

```python
grid_search = GridSearchCV(rf, param_grid, cv=3, scoring="roc_auc", n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_rf = grid_search.best_estimator_
```

```python
from sklearn.metrics import classification_report, roc_auc_score, precision_recall_curve
y_probs = best_rf.predict_proba(X_test)[:, 1]
```

```python
print("AUC-ROC Score:", roc_auc_score(y_test, y_probs))
```

```python
precision, recall, thresholds = precision_recall_curve(y_test, y_probs)
new_threshold = 0.3
y_pred_new = (y_probs > new_threshold).astype(int)
```

```python
print(classification_report(y_test, y_pred_new))
```
